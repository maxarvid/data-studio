```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler

%matplotlib inline
```

```python
df = pd.read_csv('data_privacy.csv')
df_all = pd.read_csv('data_final.csv')
df.head(2)
```

```python
plt.rcdefaults()
plt.rcParams.update({
  'patch.linewidth': 0.5,
  'patch.facecolor': '#348ABD',
  'patch.edgecolor': '#EEEEEE',
  'patch.antialiased': True,
  'font.size': 10.0,
  'axes.facecolor': '#E5E5E5',
  'axes.edgecolor': 'white',
  'axes.linewidth': '1',
  'axes.grid': True,
  'axes.titlesize': 'x-large',
  'axes.labelsize': 'large',
  'axes.labelcolor': '#555555',
  'axes.axisbelow': True,
  'axes.prop_cycle': cycler('color', ['#E24A33', '#348ABD', '#988ED5', '#777777', '#FBC15E', '#8EBA42', '#FFB5B8']),
  'xtick.color': '#555555',
  'xtick.direction': 'out',
  'ytick.color': '#555555',
  'ytick.direction': 'out',
  'grid.color': 'white',
  'grid.linestyle': '-',
  'figure.facecolor': 'white',
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})
```

```python
ax = df_all.Response.value_counts().sort_values().plot(kind='bar', figsize=(5,6), title='Privacy recommendations')
ax.grid(False)
ax.set_xticklabels( ['Noted','Supported'], rotation=0 ) 
plt.savefig("privacy_recs.pdf")

plt.show(ax)
```

```python
ax = df.Response.value_counts().sort_values().plot(kind='bar', figsize=(5,6), title='All recommendations')
ax.grid(False)
ax.set_xticklabels( ['Noted','Supported'], rotation=0 ) 
plt.savefig("all_recs.pdf")

plt.show(ax)
```

```python
ax = df.SuR.value_counts().head().sort_values().plot(kind='barh', title='Top 5 privacy recommendations recipients')
ax.grid(False, axis='y')
plt.savefig("top_recips.pdf")

plt.show(ax)
```

```python
df_privacy_plot = pd.read_csv('privacy_plot.csv')
df_privacy_plot.shape
```

```python
ax = df_privacy_plot.sort_values(by='SuR', ascending=False).plot(x='SuR', y=['Supported', 'Noted'], kind='barh', stacked=False, figsize=(10,8), title='Privacy recommendations')
ax.grid(False, axis='y')
ax.set_ylabel('')
plt.savefig("recs_reaction.pdf")

plt.show(ax)
```
