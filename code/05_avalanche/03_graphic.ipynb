```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler

%matplotlib inline

plt.rcParams.update({
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
})
```

```python
df = pd.read_csv('avalanche_with_tdidf.csv')
```

```python
df.head(2)
```

```python
df.plot(kind='scatter', x='year', y='similarity_magnitude')
```

```python
ax = df.plot(kind='scatter', x='year', y='similarity_magnitude', figsize=(15,8))
ax.set_xticks(df.year)
```

```python
df.plot(kind='scatter', x='year', y='most_similar_song_year', s=df.similarity_magnitude*1000, figsize=(15,15), alpha=.7)
ax.set
```

```python
ax = df.sort_index(ascending=False).plot(kind='barh', x='title', y='similarity_magnitude', figsize=(10,25))
ax2 = ax.twinx()
ax.yaxis.set_ticks_position('both')

plt.savefig("track_list.pdf")
```

```python
ax = df.sort_index(ascending=False).plot(kind='barh', x='most_similar_song', y='similarity_magnitude', figsize=(10,25))
ax2 = ax.twinx()
ax.yaxis.set_ticks_position('both')

plt.savefig("track_list_similar.pdf")
```

# Changing df to albums.

```python
df_album = pd.read_csv('albums.csv')
df_album
```

```python
df_album.sort_values(by='similarity_magnitude', ascending=False)
```

```python
df_album.sort_index(ascending=False).plot(kind='barh', x='album', y='similarity_magnitude', color='blue')

plt.savefig('albums.pdf')
```

```python
df_album.sort_index(ascending=False).plot(kind='barh', x='most_similar_album', y='similarity_magnitude', color='blue')

plt.savefig('albums_similar.pdf')
```

```python
ax = df_album.plot(kind='scatter', x='year', y='most_similar_album_year', s=df_album.similarity_magnitude*1000, figsize=(15,15), alpha=.7)
ax.set_xticks(df_album.year)
ax.set_yticks(df_album.most_similar_album_year)



plt.savefig('album_timeline.pdf')
```
