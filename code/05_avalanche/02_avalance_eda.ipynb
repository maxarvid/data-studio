```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import math
```

```python
df = pd.read_csv('avalanche_cleaned.csv')
df.head()
```

```python
print(df.lyrics[3])
```

```python
def tokenize(s):
    blob = TextBlob(s.lower())
    words = [token for token in blob.words if len(token)>2]
    return words
```

```python
vectorizer = TfidfVectorizer(stop_words='english', tokenizer=tokenize)

matrix = vectorizer.fit_transform(df.lyrics)

tfidf = pd.DataFrame(matrix.toarray(), columns=vectorizer.get_feature_names())
tfidf
```

```python
def print_sorted_vector(v):
    sorted_list = sorted(v.items(), key=lambda x: (x[1],x[0]), reverse=True) 
    sorted_list = sorted_list[:20]
    print('\n'.join([str(x) for x in sorted_list]))
```

```python
print_sorted_vector(tfidf.iloc[20])
```

```python
total = tfidf.sum(axis=0)
```

```python
print_sorted_vector(total)
```

```python
def doc_distance(a_vec,b_vec):
    similarity = a_vec.dot(b_vec)
    return 1-similarity

def dij(i,j):
    return doc_distance(tfidf.iloc[i], tfidf.iloc[j])
```

```python
dij(0,9)
```

```python
query_doc = 4

closest = [(i, dij(query_doc, i)) for i in range(len(tfidf))]

closest.sort(key=lambda x: x[1])

closest[:5]
```

```python
df.iloc[4]
```

```python
def most_similar_song(song_name):
    query_song = df[df.title == song_name].index[0]
    closest = [(i, dij(query_song, i)) for i in range(len(tfidf))]
    closest.sort(key=lambda x: x[1])
    return df['title'].iloc[closest[1][0]]
```

```python
most_similar_song('Suzanne')
```

## Creating new column with corresponding most similar songs

```python
df['most_similar_song'] = df.title.apply(most_similar_song)
```

```python
df.head()
```

```python
df[df.title == "Be For Real"].iloc[0,3]
```

```python
def most_similar_song_album(song_name):
    return df[df.title == song_name].iloc[0,0]

def most_similar_song_year(song_name):
    return df[df.title == song_name].iloc[0,3]
```

```python
df['most_similar_song_album'] = df.most_similar_song.apply(most_similar_song_album)
df['most_similar_song_year'] = df.most_similar_song.apply(most_similar_song_year)
df.shape
```

```python
df.tail()
```

```python
def similarity_magnitude(song_name):
    query_song = df[df.title == song_name].index[0]
    closest = [(i, dij(query_song, i)) for i in range(len(tfidf))]
    closest.sort(key=lambda x: x[1])
    return closest[1][1]
```

```python
similarity_magnitude('Suzanne')
```

```python
df['similarity_magnitude'] = df.title.apply(similarity_magnitude)
df.head()
```

```python
df.sort_values('similarity_magnitude')
```

```python
df['disimilarity_magnitude'] = df.similarity_magnitude
```

```python
df['similarity_magnitude'] = (df.disimilarity_magnitude - 1).abs()
```

```python
df.sort_values(by='similarity_magnitude', ascending=False).head()
```

```python
df.to_csv('avalanche_with_tdidf.csv', index=False)
```
