```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

import glob
```

```python
filenames = glob.glob('csvs/*2017*csv')
```

```python
filenames
```

I've scoured the files, and it looks like the files with `JC-` is additional data (the files are much smaller), so we'll be useing both.

```python
dataframes = []

for filename in filenames:
    df = pd.read_csv(filename)
    df['filename'] = filename
    dataframes.append(df)
len(dataframes)
```

```python
#Still super slow, but much better than the omnibus notebook.
everything = pd.concat(dataframes, ignore_index=True, sort=True)
everything.head()
```

```python
everything.columns
```

```python
#The JC- files have differently named columns. Duh. Let's fix that:
everything['Birth Year'].fillna(everything['birth year'], inplace=True)
```

```python
everything['Bike ID'].fillna(everything.bikeid, inplace=True)
everything['End Station ID'].fillna(everything['end station id'], inplace=True)
everything['End Station Latitude'].fillna(everything['end station latitude'], inplace=True)
everything['End Station Longitude'].fillna(everything['end station longitude'], inplace=True)
everything['End Station Name'].fillna(everything['end station name'], inplace=True)
everything['Gender'].fillna(everything['gender'], inplace=True)
everything['Start Station ID'].fillna(everything['start station id'], inplace=True)
everything['Start Station Latitude'].fillna(everything['start station latitude'], inplace=True)
everything['Start Station Longitude'].fillna(everything['start station longitude'], inplace=True)
everything['Start Station Name'].fillna(everything['start station name'], inplace=True)
everything['Start Time'].fillna(everything['starttime'], inplace=True)
everything['Stop Time'].fillna(everything['stoptime'], inplace=True)
everything['Trip Duration'].fillna(everything['tripduration'], inplace=True)
everything['User Type'].fillna(everything['usertype'], inplace=True)
```

```python
everything.head(5)
```

```python
everything.drop(columns=['bikeid', 'birth year', 'end station id',
       'end station latitude', 'end station longitude', 'end station name',
       'filename', 'gender', 'start station id', 'start station latitude',
       'start station longitude', 'start station name', 'starttime',
       'stoptime', 'tripduration', 'usertype'], inplace=True)
```

```python
#Okay! Let's save that to a csv and do some EDA (in a new notebook).
everything.to_csv('bikeshare2017.csv', index=False)
```
