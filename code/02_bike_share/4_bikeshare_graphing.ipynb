```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
phil = pd.read_csv('phil.csv')
phil.head()
```

```python
phil.dtypes
```

```python
phil['Stop Time'] = pd.to_datetime(phil['Stop Time'])
phil['Start Time'] = pd.to_datetime(phil['Start Time'])
phil.at_rest = pd.to_timedelta(phil.at_rest)
```

# Phil's busy year.

```python
phil.resample('d', on='Start Time')['Start Time'].count().plot(kind='bar', color='blue')
```

```python
phil.groupby(['Gender', 'User Type'])['Trip Duration'].sum().plot(kind='bar')
```

```python
phil.resample('W', on='Start Time')['Trip Duration'].sum().plot(kind='bar', color='blue')
```

```python
phil.resample('W', on='Start Time')['at_rest'].sum().plot(kind='bar', color='blue')
```
