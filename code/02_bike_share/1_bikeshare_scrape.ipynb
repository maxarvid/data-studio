NB: There was one file in the directory I had to delete as it was a compilation of data contained in other files. So 95 files at the time of 16 July 2018 is correct.

```python
dataframes = []

for filename in csv_names:
    df = pd.read_csv(filename)
    df['filename'] = filename
    dataframes.append(df)
len(dataframes)
```

```python
everything = pd.concat(dataframes, ignore_index=True, sort=True)
everything.tail(5)
```

```python
everything.bikeid.fillna(everything['Bike ID'], inplace=True)
```
